```python
import numpy as np
from types import SimpleNamespace
from scipy import optimize 
from scipy.optimize import minimize

%load_ext autoreload
%autoreload 2

from inauguralproject import ExchangeEconomyClass
model = ExchangeEconomyClass()
par = model.par

class ExchangeEconomyClass:

    def __init__(self):

        par = self.par = SimpleNamespace()

        # a. preferences
        par.alpha = 1/3
        par.beta = 2/3

        # b. endowments
        par.w1A = 0.8
        par.w2A = 0.3
        par.w1B = 1-par.w1A
        par.w2B = 1-par.w2A

    def utility_A(self,x1A,x2A):
        par = self.par
        u = x1A**par.alpha*x2A**(1-par.alpha)
        return u


    def utility_B(self,x1B,x2B):
        par = self.par
        u = x1B**par.beta*x2B**(1-par.beta)
        return u

    def demand_A(self,p1):
        par = self.par
        dem_A1 = par.alpha*(p1*par.w1A+par.w2A)/p1
        dem_A2 = (1-par.alpha)*(p1*par.w1A+par.w2A)
        return dem_A1 , dem_A2

    def demand_B(self,p1):
        par = self.par
        dem_B1 = par.beta*(p1*par.w1B+par.w2B)/p1
        dem_B2 = (1-par.beta)*(p1*par.w1B+par.w2B)
        return dem_B1 , dem_B2
      

    def check_market_clearing(self,p1):

        par = self.par

        x1A,x2A = self.demand_A(p1)
        x1B,x2B = self.demand_B(p1)

        eps1 = x1A-par.w1A + x1B-(1-par.w1A)
        eps2 = x2A-par.w2A + x2B-(1-par.w2A)

        return eps1,eps2




# Define utility function for Consumer A
def new_utility_A(xA):
    x1A, x2A = xA
    return -(x1A ** par.alpha * x2A ** (1 - par.alpha))

# Define objective function to maximize Consumer A's utility
def objective(p1):
    w_B = [par.w1B , par.w2B]
    x1B, x2B = demand_B(p1)
    return new_utility_A((1 - x1B, 1 - x2B), a)

# Initial guess for price
initial_price = [1 , 1]

# Bounds for price
bounds = ((0, None), (0, None))

# Solve the optimization problem
result = optimize.minimize(objective, initial_price, bounds=bounds, method='SLSQP', constraints={'type': 'ineq', 'fun': lambda p1: model.check_market_clearing(p1)})

# Extract optimal prices
optimal_prices = result.x
optimal_prices_str = ', '.join([f'{p:.2f}' for p in optimal_prices])

print("Optimal prices:", optimal_prices_str)

# Calculate optimal allocation
optimal_allocation_A = demand_A(optimal_prices, [w1_A, w2_A])
optimal_allocation_B = demand_B(optimal_prices, [1 - w1_A, 1 - w2_A])

print("\nOptimal allocation for Consumer A:")
print(f"x1_A: {optimal_allocation_A[0]:.3f}")
print(f"x2_A: {optimal_allocation_A[1]:.3f}")

print("\nOptimal allocation for Consumer B:")
print(f"x1_B: {optimal_allocation_B[0]:.3f}")
print(f"x2_B: {optimal_allocation_B[1]:.3f}")
```
